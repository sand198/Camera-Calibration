Distortion Calibration

```python
%reload_ext autoreload
%autoreload 2

import numpy as np
import cv2

from matplotlib import pylab as plt
```

#Load the numpy file that we got after running the distortion analysis

```python
# Load the data from the .npy file
loaded_data = np.load('NPY/distortion_calibration.npy', allow_pickle=True).item()
```

```python
# Extract the camera matrix and distortion coefficients
cmat = loaded_data['CameraMatrix']
distcoefs = loaded_data['DistortionCoeffs']
```

```python
# Load an image to be undistorted
image = cv2.imread("../data/chessboard/distorted.JPG")

# Make a copy of the image before undistorting
distorted_image = image.copy()
```

```python
# Undistort the image
undistorted_image = cv2.undistort(distorted_image, cmat, distcoefs)
```

```python
# Display both images side by side using matplotlib
plt.figure(figsize=(12, 6))

# Display the distorted image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(distorted_image, cv2.COLOR_BGR2RGB))
plt.title('Distorted Image')
plt.axis('off')

# Display the undistorted image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(undistorted_image, cv2.COLOR_BGR2RGB))
plt.title('Undistorted Image')
plt.axis('off')

plt.show()
```
