## Diameter of Circle in Pixel measurement. In this research we focused only on 1cm red color circle as the size quiet resembles with the size of Sika deer spot.

### Dependencies and python file where all the functions are saved

```python
# Import the necessary functions from misc
from diameter import load_pair, diameter_from_image, normalize, subtract_mask, plot_side_by_side, create_circle_colormap, process_and_display_image
import numpy as np
import matplotlib.pyplot as plt
import os
import numpy as np
```

### Load the image and mask located in the path

```python
# Define the paths to your image and mask
input_image = "./data/images/image1.jpg"
mask_path = "./data/templates/second_masked_template.png"    #Here we have two sample masked template, here in this research we only focus on 1 cm circle sized diameter# 

# Load image and mask pair
mask, image = load_pair(input_image, mask_path)

# Plot side-by-side original image and masked image
plot_side_by_side(image, image * np.expand_dims(mask, axis=2))
```

### Normalize and substract the mask

```python
# Normalize and subtract mask
image_norm = normalize(image)
image_sub = subtract_mask(image_norm, mask)

# Plot side-by-side normalized and subtracted mask images
plot_side_by_side(image_norm, image_sub)
```

```python
# Process and display the image with the custom colormap
process_and_display_image(input_image, mask_path)
```

### Diameter measurement from the area of circle in pixel.

```python
# Get diameters from image
DIAMETER, labels = diameter_from_image(image, mask, SM=1)
# Saving the diameter with labels in numpy
np.save('camera_calibration/NPY/diameter.npy', {'DIAMETER': DIAMETER, 'labels': labels})
```
